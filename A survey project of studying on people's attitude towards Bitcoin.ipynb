# A survey project of studying on people's attitude towards Bitcoin

## Research topic and background

Bitcoin is the first distributed cryptocurrency. Zhang (2021) reported the vast popularity of bitcoin and a consistent growing awareness ratio on a global scale in recent years. Moreover, the report described the wide fluctuation in Bitcoin's price since its first physical transaction of exchanging 10,000 mined Bitcoin for two pizzas in Florida, May 2010. The volatile price in Bitcoin attracts broad awareness. Many studies tried to use machine learning to forecast price trends. However, many factors drive Bitcoin price, such as economics and technology(Cai et al, 2022). Many studies tried to forecast bitcoin prices with a single driver, such as (Panagiotidis, Stengos and Vravosinos, 2019). However, such approaches overlooked the complexity of multi-driver impact. Cai et al (2022) reported a risk scenario-based approach and found that policy risks posed the severest impact on Bitcoin price. 

However, the policies announced on Bitcoin varies from country to country. El Salvador became the first country to use Bitcoin as legal tender in Sep 2021(PÃ©rez, 2021). Legal tender is defined as a form of money that courts of law are required to recognise as satisfactory payment for any monetary debt (Wikipedia, 2022). Hence, the acceptance of legal tender money is imposed by-laws(Abu-Bakar, 2018). On the other side, John, Shen and Wilson (2021) reported that China issued a blanket ban on all cryptocurrencies to root out cryptocurrency activities. More countries are found in different attitudes towards Bitcoin. Finland allows and imposes taxes on Bitcoin transactions, while Estonia concludes neither regulations nor obstacles on the usage of Bitcoin (Wikipedia, 2022). 

Many studies tried to evaluate Bitcoin as a legal currency. Baur, Hong and Lee (2018) concluded Bitcoin as a speculative investment and not as an alternative currency and medium of exchange. Notetheless, Hui et al (2020) concluded that Bitcoin behaved as currency between fiat money and a crypto-commodity in the monetary modelling. Fiat money is legal tender that is not backed by any commodity (Wikipedia, 2022). 

Bitcoin legality is probably the cornerstone where many studies and policies are played. This project is motivated to investigate the trend of Bitcoin becoming legal tender. By declaring an item legal tender, sufficient demand might be generated to determine the medium of exchange (Luther, 2019). Howarth (2022) reported that more than 81 million people created unique Bitcoin wallets on Blockchain.com as of March 2022, an increase of 72% from the previous year. The sharp increase in Bitcoin demand might suggest a strong voice in the population for the legality of Bitcoin. This project's scope is to analyse how Bitcoin is supported as a legal tender in population. The study may provide insights on areas of studies on price prediction and digital currency. Bitcoin investment is not in the scope of this project. However, the study could provide insights to understand and direct Bitcoin investment behaviours.  

The following research questions were addressed in this project which could be answered by relative objectives: 

What are people's attitudes to grant Bitcoin as legal tender? 
The objective is to explore the population's attitude on Bitcoin legality by collecting and analysing people's voting data to grant Bitcoin as legal tender, which mapped to survey question "7.	Do you support or against to grant Bitcoin as legal tender in your country?" 

How do different backgrounds impact people voting options? The objective is to evaluate the sample's representativeness to the population and assess demographic features' association with people's attitudes, which mapped to demographic survey questions including age, gender, education level and salary level. Bitcoin awareness was also checked in these survey questions.  

How do bitcoin features contribute to people's votes?
The objective is to evaluate the importance of Bitcoin features to Bitcoin legality. This evaluation was done by gathering voters' ratings on how each bitcoin feature contributed to their votes with the Likert scale type questions. Features investigated in this project were reviewed in the report of (Zhang, 2021)

This report included five sections. The following section is data collection and survey design, followed by data analysis. The subsequent section explored the Bitcoin legal status worldwide from public data. Finally, a conclusion section summarised all the findings.

## Data collection and survey design

### Survey data collection

This project used a quantitative approach to explore Bitcoin legality acceptance in the population by a public survey method. The project scope included a survey on Bitcoin legality acceptance by the sample data and public data describing existing Bitcoin legality status across countries. The scope of the survey included exploring respondents' attitudes to Bitcoin as legal tender and factors that may influence these attitudes. Factors could come from both respondents' backgrounds and Bitcoin's features. Research questions and objectives are set in accordance with the scope as above.



### Survey design

Type of questions An assorted type of questions, including multiple-choice questions and open-ended questions, were adopted in this survey. A block-level survey flow was adopted in this project, where each research question formed a block (as below). A pre-test was run before the release of the survey. A final number of 71 responses was collected.

In this project, a public survey method is used. The survey was structured as below.

Introduction.

Survey questions on demographic features or backgrounds.

Survey questions on voters' awareness and stance on Bitcoin.

Survey questions on how Bitcoin typical features impact participants' votes.

### Public data collection

People's attitude on Bitcoin's legality could be compared with the actual Bitcoin legal status across countries to understand gaps or trends. Wikipedia is a free encyclopedia and the largest and most popular reference website in the world(Ayers, Matthews and Yates, 2008). Wikipedia platform consistently publishes up-to-date data based on information collected from the public. In this project, data were collected from Wikipedia articles. However, the list of countries is incomplete. As of Mar 2022, there are ninety-two countries listed on the website with details of Bitcoin legality status traceable references. There are five categories built accordingly, Legal tender, Permissive(it's legal to use bitcoin), Contentious(some legal restrictions on usage of bitcoin), Contentious(interpretation of old laws, but bitcoin is not prohibited directly), Hostile (full or partial prohibition) and No data. The contentious categories will be consolidated as one category in this project. Each category is appropriately mapped with a unique colour on the map. However, these categories are not explicitly described in line with the content. Hence, this project manually traced different records to match each country to a category by cross-checking the description and the colour map. Additionally, more countries are not listed as the information is not available. Dadax Limited (2022) reported a total of 195 countries in the world. Hence, the project imputed unlisted countries by Wikipedia as dummies and grouped them under the 'No Data' category.

### Sample and Population: 

It may sound reasonable to include all adults all over the world as long as they make transactions with currency. However, due to the time limit and survey's reach in scale, this survey only targeted the adult English speakers worldwide as the survey came in the form of English language. There is a probable bias that the survey sample is not representing the population of different languages. However, that is not within our project scope. The sampling tried to reach a wider population as possible, including the UOL students, surveyor's social networks and so on. Different types of popular forums will be included to increase sample randomness. However, the starting point of the survey would be closely related to the surveyor's social network. Hence, each participant was requested to invite two more participants to the survey. Hence, the survey could be randomly distributed without the control of the surveyor. Seventy-one responses were collected. The response rate was higher than expected due to wide invitations to social networks.

After that, detailed analysises were made on each of the feature to understand how it contribute to different voter groups.

The likert scale will be discussed as three categories, namely likely/verylikely, neutral, unlikely/very unlikely. When voters chose neutral in the likert scale, it will bereferred as irrelevant or unimportant of this feature.

### Correlation among different features

```python
#encode categorical features with numbers.

df1 = df.drop('age', axis=1)

df1 = df1.apply(lambda x: x.cat.codes)
df1['age'] = df['age']
```

If we split the association into different scales as following: 

0.00 as None association

0.01 ~ 0.09 as Negligible association

0.10 ~ 0.29 as Weak association

0.30 ~ 0.59 as Moderate association

0.60 and above as strong association

1.00 as perfect association



Most variables pairs have negligible or weak associations while variables including "gender", "age", "incomeLeve", and "educationLevel" have negligible or weak associations to most of the variables. However, "age" variable has moderate association with "educationLevel" and "riskyDigitWallet". Further exploration was made as below to visualize the associations.

On the other side, Most Bitcoin features are weakly or moderately associated with each other. This is understandable as all features reflect different aspects of Bitcoin. 

Overall, there is no dominant variable that strongly correlates with the votes. All variables have a weak or negligible association with the votes.

## Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
